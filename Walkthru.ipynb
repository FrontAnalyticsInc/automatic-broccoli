````python
import auto_broccoli as broc
````

Parameters you can pass in here are:
```
 - df, (str path to CSV)
 - categorical_as_ints=False, (if a column of ints is categorical, set to True)
 - only_significant=False,    (only output statistically significant results, dependent on sig_level)
 - sig_level=0.05,           
 - min_samples=30,            (for statisical soundness)
 - specific_config=None,      (If you want to override the configuration in the config file)
 - table_name=None            (str, the table name)
```

### On it's own, it generates a sample dataset

````python
# Fake generated data
ab = broc.AutoBroccoli()
````

````python
ab.df.head()
````

````python
resultsdf = ab.main()
````

````python
# Insights / results
resultsdf
````

### Or you can pass it a path to a CSV

````python
ab = broc.AutoBroccoli(df='data/bulk_yt_sample.csv', only_significant=True)
````

````python
resultsdf = ab.main()
````
